Problem 01: Implement Bresenham's Circle Drawing Algorithm to calculate and plot the pixel coordinates for a circle centered at (0,0) with a radius of 100.



```python
import matplotlib.pyplot as plt

def draw_circle_bresenham(xc, yc, r):
    x = 0
    y = r
    d = 3 - 2 * r
    points = []

    while x <= y:

        points.extend([
            (xc + x, yc + y),
            (xc - x, yc + y),
            (xc + x, yc - y),
            (xc - x, yc - y),
            (xc + y, yc + x),
            (xc - y, yc + x),
            (xc + y, yc - x),
            (xc - y, yc - x)
        ])

        if d < 0:
            d = d + 4 * x + 6
        else:
            d = d + 4 * (x - y) + 10
            y -= 1
        x += 1

    return points

def plot_circle(points, title):
    x_values, y_values = zip(*points)
    plt.scatter(x_values, y_values, color='blue', marker='*')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.gca().set_aspect('equal')
    plt.grid(True)
    plt.show()

# Example usage
xc, yc = 0, 0
radius = 100


circle_points = draw_circle_bresenham(xc, yc, radius)
plot_circle(circle_points, "Bresenham's Circle Drawing Algorithm")
```

Problem 02: Implement Bresenham's line drawing algorithm to draw a line from point (9,18) to point (14, 22) .

```python
import matplotlib.pyplot as plt

def draw_line_DDA(x1, y1, x2, y2):
    points = []
    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    for _ in range(steps + 1):
        points.append((round(x), round(y)))
        x += x_inc
        y += y_inc

    return points

def plot_line(points, title):
    x_values, y_values = zip(*points)
    plt.plot(x_values, y_values, marker='*',color='g')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

# Example usage:
x1,y1 = 9, 18
x2,y2 =14, 22
#Bresenham'algorithm
bresenham_points = draw_line_DDA(x1, y1, x2, y2)
plot_line(bresenham_points, 'Bresenham\'s Line Drawing Algorithm')

```

Problem 03: Implement the Digital Differential Analyzer (DDA) algorithm to draw a line from point (2, 3) to point (9, 8) .

```python
import matplotlib.pyplot as plt
def draw_line_DDA(x1, y1, x2, y2):
    points = []
    dx = x2 - x1
    dy = y2 - y1
    steps = max(abs(dx), abs(dy))

    x_increment = dx / steps
    y_increment = dy / steps

    x, y = x1, y1

    for _ in range(steps):
        points.append((round(x), round(y)))
        x += x_increment
        y += y_increment

    return points

def plot_line(points, title):
    x_values, y_values = zip(*points)
    plt.plot(x_values, y_values, marker='o')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

x1, y1 = 2,3
x2, y2 = 9,8

# DDA algorithm
dda_points = draw_line_DDA(x1, y1, x2, y2)
plot_line(dda_points, 'DDA Line Drawing')
```

Problem 04: Implement the Midpoint Circle Drawing Algorithm to draw a circle with center at (0, 0) and radius 5 .

```python
import matplotlib.pyplot as plt

def midpoint_circle(xc, yc, r):
    x = 0
    y = r
    p = 1 - r   # initial decision parameter

    points = []

    while x <= y:
        # 8-way symmetry
        points.extend([
            (xc + x, yc + y),
            (xc - x, yc + y),
            (xc + x, yc - y),
            (xc - x, yc - y),
            (xc + y, yc + x),
            (xc - y, yc + x),
            (xc + y, yc - x),
            (xc - y, yc - x),
        ])

        if p < 0:
            p = p + 2 * x + 3
        else:
            p = p + 2 * (x - y) + 5
            y -= 1
        x += 1

    return points


#  Draw circle with center (0,0) and radius 5
xc, yc, r = 0, 0, 5
circle_points = midpoint_circle(xc, yc, r)

# Plot points
for point in circle_points:
    plt.plot(point[0], point[1], "ro")


plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Midpoint Circle Drawing Algorithm")
plt.show()
```

Problem 05: Implement the Midpoint Line Drawing Algorithm to draw a line from point (10,18) to point (15,20).

```python
import matplotlib.pyplot as plt



def draw_line_midpoint(x1, y1, x2, y2):


    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    d=dy-(dx/2)

    x = x1
    y = y1
    points = []
    points.append((x, y))

    while x<x2:
       x+=1
       if d<0:
           d=d+dy
       else:
           d=d+dy-dx
           y+=1
       points.append((x, y))


    return points

def plot_line(points, title):
    x_values, y_values = zip(*points)
    plt.plot(x_values, y_values, marker='*', color='g')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

# Example usage:
x1, y1 = 10,18
x2, y2 = 15,20


midpoint_points = draw_line_midpoint(x1, y1, x2, y2)
plot_line(midpoint_points, 'Midpoint Line Drawing Algorithm')
```

Problem 06: Implement a 2D translation of a triangle with vertices at (1, 1), (2, 3), and (3, 1) using a translation matrix to shift it by 2 units in both x and y directions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Original triangle points A(1,1), B(3,1), C(2,4)
original_points = np.array([
    [1, 1],  # A
    [2, 3],  # B
    [3, 1],  # C
    [1, 1]   # Close the triangle
])

# Translation distances
tx = 2   # Translate 2 units in x-direction
ty = 2   # Translate 3 units in y-direction

# Translation matrix (just add tx, ty to each point)
translation_vector = np.array([tx, ty])
translated_points = original_points + translation_vector

# Plot original and translated triangle
plt.figure(figsize=(6, 6))
plt.plot(original_points[:, 0], original_points[:, 1], 'bo-', label='Original Triangle ABC')
plt.plot(translated_points[:, 0], translated_points[:, 1], 'go-', label='Translated Triangle A\'B\'C\'')

plt.title('2D Translation of Triangle ABC')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

Problem 07: Implement a 2D reflection of a triangle with vertices at (1, 1), (3, 1), and (2, 4) about both the x-axis and y-axis using reflection matrices.

```python
import numpy as np
import matplotlib.pyplot as plt

# Triangle points A(1,1), B(3,1), C(2,4)
original_points = np.array([
    [1, 1],  # A
    [3, 1],  # B
    [2, 4],  # C
    [1, 1]   # To close the triangle
])

# Choose reflection type: 'x', 'y', 'origin', or 'y=x'
reflection_type = 'x'  # Change this to 'y', 'origin', 'y=x' as needed

# Define reflection matrices
reflection_matrices = {
    'x': np.array([[1, 0], [0, -1]]),       # Reflect over X-axis
    'y': np.array([[-1, 0], [0, 1]]),       # Reflect over Y-axis
    'origin': np.array([[-1, 0], [0, -1]]), # Reflect over origin
    'y=x': np.array([[0, 1], [1, 0]])       # Reflect over y = x line
}

# Apply reflection
reflection_matrix = reflection_matrices[reflection_type]
reflected_points = original_points @ reflection_matrix.T

# Plotting
plt.figure(figsize=(6, 6))
plt.plot(original_points[:, 0], original_points[:, 1], 'bo-', label='Original Triangle ABC')
plt.plot(reflected_points[:, 0], reflected_points[:, 1], 'r*-', label=f'Reflected over {reflection_type}')

plt.title(f'2D Reflection over {reflection_type.upper()}')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

Problem 08: Implement a 2D shearing transformation of a triangle with vertices at (1, 1), (2, 3), and (3, 1) using a shearing matrix with shx=0.5 for x-direction and shy=0.3 for y-direction.

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_triangle(p, c, lbl):
    plt.plot([p[0][0], p[1][0], p[2][0], p[0][0]],
             [p[0][1], p[1][1], p[2][1], p[0][1]], c, label=lbl)

def shear(p, shx=0, shy=0):
    m = np.array([[1, shx], [shy, 1]])
    return p @ m.T

# Define the original triangle vertices
tri = np.array([[1,1],[2,3],[3,1]])

# Shearing
tri_x = shear(tri, shx=0.5)
tri_y = shear(tri, shy=0.3)

plt.figure()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
draw_triangle(tri, 'b', 'Original')
draw_triangle(tri_x, 'r', 'Sheared X (0.5)')
draw_triangle(tri_y, 'g', 'Sheared Y (0.3)')
plt.axhline(0, color='k', alpha=0.3)
plt.axvline(0, color='k', alpha=0.3)
plt.legend(); plt.axis('equal')
plt.show()
```

Problem 09: Implement a 2D scaling transformation for a triangle with vertices at (1,1), (3,1), and (2,4) using a scaling matrix with factors sx=2 and sy=2.

Problem 10: Implement a 2D rotation of a line segment from (0, 0) to (3, 0) using a rotation matrix for angles 0°, 45°, 90°, 135°, and 180°.

```python
import numpy as np
import matplotlib.pyplot as plt

def rotate(points, angle):
    rad = np.radians(angle)
    R = np.array([[np.cos(rad), -np.sin(rad)],
                  [np.sin(rad),  np.cos(rad)]])
    return points @ R.T

# Original line (from origin along the x-axis)
line = np.array([[0, 0], [3, 0]])
angles = [0, 45, 90, 135, 180]
colors = ['b', 'r', 'g', 'm', 'c']

plt.figure(figsize=(6, 6))
plt.grid(True)

for a, c in zip(angles, colors):
    rot_line = rotate(line, a)
    plt.plot(rot_line[:,0], rot_line[:,1], c, label=f'{a}°')

plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
plt.legend()
plt.axis('equal')
plt.show()

# Optional Animation
def animate_rotation():
    plt.figure(figsize=(6, 6))
    for a in range(0, 361, 5):
        plt.clf()
        rot_line = rotate(line, a)
        plt.plot(rot_line[:,0], rot_line[:,1], 'r', label=f'{a}°')
        plt.axhline(0, color='k', lw=0.5)
        plt.axvline(0, color='k', lw=0.5)
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.legend()
        plt.axis('equal'); plt.xlim(-4,4); plt.ylim(-4,4)
        plt.pause(0.05)
    plt.show()


```

Problem 11: Implement a 2D rotation of a triangle with vertices at (1, 0), (2, 2), and (3, 0) using
a rotation matrix for angles 0°, 45°, 90°, 135°, and 180°.


```python
import numpy as np
import matplotlib.pyplot as plt

def rotate(points, angle):
    rad = np.radians(angle)
    R = np.array([[np.cos(rad), -np.sin(rad)],
                  [np.sin(rad),  np.cos(rad)]])
    return points @ R.T

#  Define the original triangle vertices
triangle = np.array([[1,0],[2,2],[3,0]])
angles = [0, 45, 90, 135, 180]
colors = ['b','r','g','m','c']

plt.figure(figsize=(6,6))
plt.grid(True)

for a,c in zip(angles,colors):
    rot_tri = rotate(triangle,a)
    pts = np.vstack([rot_tri, rot_tri[0]])
    plt.plot(pts[:,0], pts[:,1], c, label=f'{a}°')

plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.axis('equal'); plt.xlim(-4,4); plt.ylim(-4,4)
plt.title("2D Rotation of a Triangle")
plt.show()
```
